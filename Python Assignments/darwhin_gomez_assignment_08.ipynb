```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Lets Load the data

```python
file_path = 'pokemon.csv'  # Replace with the correct path if needed
data = pd.read_csv(file_path)
```

Data Exploration

```python
print("Column Information:")
print(data.info())
```

```python
print("First few rows of the dataset:")
print(data.head())
print("\nSummary statistics:")
print(data.describe())
print("\nMissing values:")
print(data.isnull().sum())
```

I see that type two has some nas, i would like to replace that with an object we will call None

Data Wrangling
Standardize Cols Names to lower_case

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')

# Verify the changes
print("Updated Column Names:")
print(data.columns)
```

```python
data['type_2'] = data['type_2'].fillna("None")

# Verify the changes
print(data['type_2'].value_counts())
```

Lets make up a column and call it total_power
we will add hp attack and defense and multiply by 10

```python
data['total_power'] = (data['hp'] + data['attack'] + data['defense']) * 10

# Verify the new column
print(data[['hp', 'attack', 'defense', 'total_power']].head())
```

lets make generation a level

```python
data['generation'] = data['generation'].astype('object')

# Verify the column type
print(data.dtypes['generation'])
```

Lets make some visualizations

```python
# Select numeric columns to plot
numeric_columns =[col for col in data.select_dtypes(include=['int64', 'float64']).columns if col != '#']


# Plot distributions for each numeric column grouped by `type_1`
for col in numeric_columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=data, x='type_1', y=col, palette='Set3')
    plt.title(f'Boxplot of {col} by Type 1', fontsize=16)
    plt.xlabel('Type 1', fontsize=12)
    plt.ylabel(col, fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()
```

```python
# Plot a histogram for 'hp'
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='hp', kde=True, bins=20, color='blue')
plt.title('Distribution of HP', fontsize=16)
plt.xlabel('HP', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()

# Plot a histogram for 'attack'
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='attack', kde=True, bins=20, color='green')
plt.title('Distribution of Attack', fontsize=16)
plt.xlabel('Attack', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()

# Plot a histogram for 'total_power'
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='total_power', kde=True, bins=20, color='orange')
plt.title('Distribution of Total Power', fontsize=16)
plt.xlabel('Total Power', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```

Part one

```python

# Create a scatter plot of speed and spc_attack with legend and style 
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=data,
    x='speed',
    y='sp._atk',
    hue='type_1',
    style='legendary',
    palette='Set2',
    s=100,  # Size of the markers
    alpha=0.8  # Transparency
)
plt.title('Speed vs Special Attack by Type and Legendary Status', fontsize=16)
plt.xlabel('Speed', fontsize=12)
plt.ylabel('Special Attack', fontsize=12)
plt.legend(title='Type & Legendary', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

The plot shows all speed and attack plotted with the color representing the main type of the pokemon and a X represent a legendary pokemon. If we take a close look we see that some of the highest values are X. Legendary indeed. This was with seaborn now lets see with matplotlib and we will do attack vs sp._atk

```python
#  first we assign colors and shapes
unique_types = data['type_1'].unique()
type_colors = {type_: plt.cm.Set2(i / len(unique_types)) for i, type_ in enumerate(unique_types)}
shape_markers = {True: 'o', False: 's'}  # Circle for legendary, square otherwise

# Create the scatter plot
plt.figure(figsize=(12, 8))
for type_ in unique_types:
    subset = data[data['type_1'] == type_]
    for legendary_status in [True, False]:
        legendary_subset = subset[subset['legendary'] == legendary_status]
        plt.scatter(
            legendary_subset['attack'],
            legendary_subset['sp._atk'],
            color=type_colors[type_],
            marker=shape_markers[legendary_status],
            label=f"{type_} ({'Legendary' if legendary_status else 'Non-Legendary'})",
            alpha=0.8,
            s=100
        )

# Add title, labels, and legend
plt.title('Attack vs Special Attack by Type and Legendary Status', fontsize=16)
plt.xlabel('Attack', fontsize=12)
plt.ylabel('Special Attack', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Type & Status')
plt.grid(True)
plt.show()
```

This is much more cluttered less intuitive and took more work to set up. But again we see legendairy pokemons have the higher values for both of these variables.

```python
# Get the top 20 Pokmon by total_power
top_20 = data.nlargest(20, 'total_power')



```

```python
# Count the number of Pokmon in each generation
generation_counts = top_20['generation'].value_counts().sort_index()
generation_counts
```

Generation 3 is has the dominates the top 20

```python
# Get the top 20 Pokmon by speed
top_20 = data.nlargest(20, 'speed')

# Create a color palette for type_1
type1_palette = sns.color_palette('viridis', len(top_20['type_1'].unique()))
type1_colors = dict(zip(top_20['type_1'].unique(), type1_palette))

# Plot using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(
    data=top_20,
    y='name',  # Pokmon names on the y-axis
    x='speed',
    hue='type_1',  # Map colors to type_1
    dodge=False,   # seperate bars
    palette=type1_colors
)
plt.title('Top 20 Pokmon by Speed', fontsize=16)
plt.xlabel('Speed', fontsize=24)
plt.ylabel('Pokmon', fontsize=24)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.legend(title='Type 1', bbox_to_anchor=(1.05, 1), loc='upper left')  # Add a legend for Type 1
plt.show()
```
