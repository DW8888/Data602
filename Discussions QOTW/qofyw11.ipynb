```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```

These questions relate to the seaborn library.

1. How do you plot a histogram in Seaborn?  


```python
# Load the dataset
data = pd.read_csv("great_castles.csv")

# Plot a histogram for a specific column
sns.histplot(data['total_population'], kde=True)  
plt.title("Histogram of Castle Heights")
plt.show()
```


2. Plot a histogram with NAs dropped.
.dropna()
3. How do you set the color for a histogram?
color="red"

```python
sns.histplot(data['guards'].dropna(), kde=True, color='orange')
plt.title("Histogram of Guards with Custom Color")
plt.xlabel("Number of Guards")
plt.show()
```

4. What type of plot would allow you to compare two continuous features?  Give an example of code.
a scatter plot

```python
sns.scatterplot(data=data, x='income', y='total_population', hue='castle_names', palette='viridis')
plt.title("Scatterplot of Guards vs Population by Lord")
plt.xlabel("Number of Guards")
plt.ylabel("Population")
plt.show()
```

5. Give example of a correlation plot.


```python
# Compute the correlation matrix
correlation_matrix = data[['total_population', 'staff', 'guards', 'income', 'debts']].corr()



plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix for Numerical Columns")
plt.show()
```

6. Change the figure size of your plot(s).

```python
plt.rcParams['figure.figsize'] = [15,12] 

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix for Numerical Columns")
plt.show()
```
