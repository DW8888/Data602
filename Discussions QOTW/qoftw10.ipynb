```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

great_castles = pd.read_csv("great_castles.csv")


plt.figure(figsize=(10, 6))
sns.barplot(x="castle_names", y="total_population", data=great_castles, color="skyblue", label="Population")
sns.barplot(x="castle_names", y="guards", data=great_castles, color="salmon", label="Troops")

# Customizing plot
plt.title("Population and Troop Count by Castle")
plt.xlabel("Castle Name")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Legend")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="castle_names", y="total_population", data=great_castles, color="skyblue", label="Population")
sns.barplot(x="castle_names", y="guards", data=great_castles, color="salmon", label="Troops")

# Customizing plot for minimalism
plt.title("Population and Troop Count by Castle", fontsize=14)
plt.xticks(rotation=45)
plt.xlabel("")
plt.ylabel("")
plt.legend(title="")  # Minimalist legend
sns.despine()  # Remove borders/spines
plt.grid(False)  # Remove gridlines

plt.show()
```
